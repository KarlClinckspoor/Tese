```python
mpl.rcParams.update({'font.size': 14})
                     
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(6,7))
for i, file in enumerate(files[::1]):
    stack_factor = 10 ** i/10
    df = pd.read_csv(file, sep='  ', names=['q', 'I'], engine='python')
    ax.plot(df['q'], df['I'] * stack_factor, label=file[9:11] + '%', linewidth=2)
  
    m_y = (df['I'] * stack_factor).max()
    pos_x = df['I'].idxmax()
    m_x = df['q'][pos_x]
    d = 2 * np.pi / m_x
    
    ax.text(m_x, 1.1*m_y, f'{d:.1f}nm', horizontalalignment='center', color=f'C{i}',
           fontdict={'weight':'bold'})
    ax.text(m_x, m_y, '|', horizontalalignment='center', verticalalignment='top', color=f'C{i}')
    
    y = df['I'][len(df)-1] * stack_factor
    x = 2.5
    ax.text(x*1.0, y * 1.5, file[9:11] + '%', horizontalalignment='right', color=f'C{i}', 
            fontdict={'weight':'bold'})

ax.set(yscale='log', xlabel='q/nm$^{-1}$', ylabel='Scattering Intensity/a.u.', xlim=(0.1, 2.5))
ax.xaxis.set_tick_params(which='major', width=1, length=10)
ax.xaxis.set_tick_params(which='minor', width=1, length=4)



# from matplotlib.ticker import LogFormatterSciNotation, LogLocator, NullFormatter
# ax.xaxis.set_minor_locator(LogLocator(subs=('all')))
# ax.xaxis.set_minor_formatter(LogFormatterSciNotation(labelOnlyBase=False, 
#                                                      minor_thresholds=(2,0.5)))
# ax.xaxis.set_major_formatter(NullFormatter())
# plt.setp(ax.xaxis.get_minorticklabels(), rotation=90 )
# #plt.setp(ax.get_xticklabels(minor=True), visible=True)

ax.yaxis.set_tick_params(which='major', width=0, length=10)
ax.yaxis.set_tick_params(which='minor', width=0, length=4)
ax.yaxis.set_ticklabels('')

# ax.set_title('CTAB 300 mmol.L$^{-1}$, % (m/m) Ur', loc='right')
# ax.legend(ncol=3, title='CTAB 300mM')
fig.tight_layout()
# plt.show()

fig.savefig(material_suplementar_path + r'\CTAB300Ur37-45-linear.png', dpi=600)
```

## CTAB 100, 200, 300 Ur 40

```python
files = glob.glob('*CTAB*40*.dat')
files = [i for i in files if ('qbaixo' not in i) and ('-2' not in i) and ('NaSal' not in i)
        and ('100a' not in i)]
files
```

```python
%matplotlib inline
mpl.rcParams.update({'font.size': 14, 'text.usetex':False})

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(9,6))
styles = ['-', '--', '-.']
styles.reverse()

for i, file in enumerate(files):
    df = pd.read_csv(file, sep=r'\s+', names=['q', 'I'], engine='python', skipinitialspace=True, header=1)
    ax.plot(df['q'], df['I'], label=file[4:7] + ' mmol.L$^{-1}$', linestyle=styles.pop(), linewidth=3)
    if file[4:7] != '100':
        m_y = df['I'].max()
        pos_x = df['I'].idxmax()
        m_x = df['q'][pos_x]
        d = 2 * np.pi / m_x
    else:
        pos_x = 10
        m_x = df['q'][pos_x]
        m_y = df['I'][pos_x]
        d = 2 * np.pi / m_x

    ax.text(m_x, 1.05*m_y, f'{d:.1f}nm', horizontalalignment='center', color=f'C{i}', 
            fontdict={'weight':'bold'})
    ax.text(m_x, m_y, '|', horizontalalignment='center', verticalalignment='top', color=f'C{i}')
    
ax.axvline(m_x, color=f'C{i}', linestyle='--')
ax.axvline(2*m_x, color=f'C{i}', linestyle='--')
ax.axvline(3*m_x, color=f'C{i}', linestyle='--')
ax.axvline(4*m_x, color=f'C{i}', linestyle='--')
ax.axvline(5*m_x, color=f'C{i}', linestyle='--')

ax.set(xscale='log', yscale='log', xlabel='q/nm$^{-1}$', 
       ylabel='Intensidade de espalhamento', xlim=(0.1, 5.5))
ax.xaxis.set_tick_params(which='major', width=1, length=10)
ax.xaxis.set_tick_params(which='minor', width=1, length=4)
ax.yaxis.set_tick_params(which='major', width=1, length=10)
ax.yaxis.set_tick_params(which='minor', width=1, length=4)
ax.legend(title='40% (m/m) ureia, C$_{16}$TAB')
#ax.set_title('', loc='right')

fig.savefig(tese_path + r'\CTAB_conc.pdf')
```

### Separados com indexação de picos

```python
from scipy.signal import find_peaks_cwt
```

```python

for i, file in enumerate(files):
    fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(9,6))
    df = pd.read_csv(file, sep=r'\s+', names=['q', 'I'], engine='python', skipinitialspace=True, header=1)
    peaks = find_peaks_cwt(df['I'], widths=[5])
    peak = peaks[1]
    ax.plot(df['q'], df['I'] / df['I'].min(), label=file)
    
    ax.axvline(df['q'][peak], color=f'C{i}', linestyle='--')
    ax.axvline(df['q'][peak]*2, color=f'C{i}', linestyle='--')
    ax.axvline(df['q'][peak]*3, color=f'C{i}', linestyle='--')
    ax.axvline(df['q'][peak]*4, color=f'C{i}', linestyle='--')
    ax.axvline(df['q'][peak]*5, color=f'C{i}', linestyle='--')
    ax.axvline(df['q'][peak]*6, color=f'C{i}', linestyle='--')
    
    
    ax.set(xscale='log', yscale='log', xlabel='q/nm', ylabel='I')
    ax.set_title('TTAB, 40% (m/m) Ur', loc='right')
    ax.legend(ncol=1)
# fig.tight_layout()
# plt.show()
# fig.savefig(r'C:\Users\Karl\Google Drive\Doutorado\Tese\imagens\saxs\CTAB300Ur37-45.pdf')
```
