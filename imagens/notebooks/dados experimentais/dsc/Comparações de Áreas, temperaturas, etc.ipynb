# Sem Salicilato, sem CTAB 400 mM

```python
mpl.rcParams.update({'font.size': 14, 'text.usetex':False, 'text.latex.unicode':True})
mpl.rcParams.update({'mathtext.fontset':'dejavusans'})

temps = pd.read_excel('Temperaturas e áreas de transição.xlsx', sheet_name='Tudo', na_values='-', header=1)
temps = temps.loc[:, '% Ureia':'Lres/°C']
temps_s = temps[temps['C Surf'] != 400]
temps_s = temps_s[temps_s['C NaSal']==0]
```

## Lres

### % ur

```python
fig, ax1 = plt.subplots(1, 1, figsize=(9,6))
fig, ax2 = plt.subplots(1, 1, figsize=(9,6))
fig, ax3 = plt.subplots(1, 1, figsize=(9,6))


mark_dict = {'CTAB':'s', 'DTAB':'o', 'TTAB':'^'}

for unique in temps['Surf'].unique():
    fig_dict = {'label':unique, 'markersize':5, 'linewidth':1}
    filtro1 = (temps_s['Surf'] == unique) & (temps_s['C Surf'] == 100)
    filtro2 = (temps_s['Surf'] == unique) & (temps_s['C Surf'] == 200)
    filtro3 = (temps_s['Surf'] == unique) & (temps_s['C Surf'] == 300)
    ax1.plot(temps_s[filtro1]['% Ureia'], temps_s[filtro1]['Lres/°C'], **fig_dict, marker=mark_dict[unique])
    ax2.plot(temps_s[filtro2]['% Ureia'], temps_s[filtro2]['Lres/°C'], **fig_dict, marker=mark_dict[unique])
    ax3.plot(temps_s[filtro3]['% Ureia'], temps_s[filtro3]['Lres/°C'], **fig_dict, marker=mark_dict[unique])
    
ax1.legend()
ax2.legend()
ax3.legend()

ax1.set(xlabel='% (m/m) ureia', ylabel='$L_{res}$/°C')
ax2.set(xlabel='% (m/m) ureia', ylabel='$L_{res}$/°C')
ax3.set(xlabel='% (m/m) ureia', ylabel='$L_{res}$/°C')
ax1.set_title('Surfactante 100 mmol.L$^{-1}$', loc='right')
ax2.set_title('Surfactante 200 mmol.L$^{-1}$', loc='right')
ax3.set_title('Surfactante 300 mmol.L$^{-1}$', loc='right')
```

```python
fig1, ax1 = plt.subplots(1, 1, figsize=(9,6))
fig2, ax2 = plt.subplots(1, 1, figsize=(9,6))
fig3, ax3 = plt.subplots(1, 1, figsize=(9,6))

mark_dict = {'CTAB':'s', 'DTAB':'o', 'TTAB':'^'}
c_dict = {'CTAB':'C0', 'DTAB':'C1', 'TTAB':'C2'}

def replace_name(string):
    if string == "CTAB":
        return r'C$_{16}$TAB'
    if string == "TTAB":
        return r'C$_{14}$TAB'
    if string == "DTAB":
        return r'C$_{12}$TAB'

for unique in temps['Surf'].unique():
    fig_dict_res = {'label':replace_name(unique) + ': Res', 'markersize':5, 'linewidth':1}
    fig_dict_aq = {'label':replace_name(unique) + ': Aq', 'markersize':5, 'linewidth':1, 
                   'linestyle':'--', 'markerfacecolor':'w'}
    filtro1 = (temps_s['Surf'] == unique) & (temps_s['C Surf'] == 100)
    filtro2 = (temps_s['Surf'] == unique) & (temps_s['C Surf'] == 200)
    filtro3 = (temps_s['Surf'] == unique) & (temps_s['C Surf'] == 300)
    
    c1 = ax1.plot(temps_s[filtro1]['% Ureia'], temps_s[filtro1]['Lres/°C'], c = c_dict[unique],
             **fig_dict_res, marker=mark_dict[unique])
    c2 = ax2.plot(temps_s[filtro2]['% Ureia'], temps_s[filtro2]['Lres/°C'], c = c_dict[unique],
             **fig_dict_res, marker=mark_dict[unique])
    c3 = ax3.plot(temps_s[filtro3]['% Ureia'], temps_s[filtro3]['Lres/°C'], c = c_dict[unique],
             **fig_dict_res, marker=mark_dict[unique])
    
    d1 = ax1.plot(temps_s[filtro1]['% Ureia'], temps_s[filtro1]['Laq/°C'], c = c_dict[unique],
             **fig_dict_aq, marker=mark_dict[unique])
    d2 = ax2.plot(temps_s[filtro2]['% Ureia'], temps_s[filtro2]['Laq/°C'], c = c_dict[unique],
             **fig_dict_aq, marker=mark_dict[unique])
    d3 = ax3.plot(temps_s[filtro3]['% Ureia'], temps_s[filtro3]['Laq/°C'], c = c_dict[unique],
             **fig_dict_aq, marker=mark_dict[unique])
    
    
#ax1.legend((*c1, *c2, *c3, *d1, *d2, *d3), ('CTAB', 'DTAB', 'TTAB', 'CTAB', 'DTAB', 'TTAB'),ncol=3)

ax1.legend(ncol=3)
ax2.legend(ncol=3)
ax3.legend(ncol=3)

l1 = [i * 1.2 for i in ax1.get_ylim()]
l2 = [i * 1.2 for i in ax2.get_ylim()]
l3 = [i * 1.2 for i in ax3.get_ylim()]

ax1.set_ylim(l1)
ax2.set_ylim(l2)
ax3.set_ylim(l3)

ax1.set(xlabel='% (m/m) ureia', ylabel='$L$/°C')
ax2.set(xlabel='% (m/m) ureia', ylabel='$L$/°C')
ax3.set(xlabel='% (m/m) ureia', ylabel='$L}$/°C')
ax1.set_title('Surfactante 100 mmol.L$^{-1}$', loc='right')
ax2.set_title('Surfactante 200 mmol.L$^{-1}$', loc='right')
ax3.set_title('Surfactante 300 mmol.L$^{-1}$', loc='right')

fig1.savefig(r'..\..\..\dsc\L_100mM_aq_res.pdf')
```
