### 100 mM

```python
def sort(x):
    if 'CTAB' in x:
        return 0
    elif 'TTAB' in x:
        return 1
    elif 'DTAB' in x:
        return 2
    
c100 = [file for file in files if '100' in file if '45%' in file]
c100.sort(key = sort)

styles = ['-', '--', '-.']
styles.reverse()

fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(6*1.5,4*1.5))
surf = re.compile('[CDT]TAB')
for file in c100:
    arq = pd.read_csv(file, encoding='ANSI', header=53, names=['a', 'T', 'Q', 'N'], sep=' ', skipfooter=1, skiprows=[9803])
    s = surf.findall(file)[0].replace('CTAB', 'C$_{16}$TAB').replace('TTAB', 'C$_{14}$TAB').replace('DTAB', 'C$_{12}$TAB')
    plt.plot(arq['T'], arq['Q'], label=s, linestyle=styles.pop(), linewidth=2.5)
    
plt.legend(title='3% Surf., 45% urea', ncol=1)
ax.set_xlabel('Temperature/°C')
ax.set_ylabel('Heat flow/W.g$^{-1}$')
#ax.set_title('100 mmol.L$^{-1}$ de surfactante em 45% m/m de ureia', loc='right')
fig.savefig(material_suplementar_path + r'\Surf_100mm_45p.png', dpi=600)
```
