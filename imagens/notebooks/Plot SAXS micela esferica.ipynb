```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp

import locale
locale.setlocale(locale.LC_ALL, '')

import matplotlib as mpl
mpl.rcParams.update({'font.size': 14, 'text.usetex':False, 'text.latex.unicode':True})
mpl.rcParams.update({'mathtext.fontset':'dejavusans'})
mpl.rcParams['axes.formatter.use_locale'] = True
```

```python
def Intensidade (q, R):
    qR = q * R
    return ( 3 * ( np.sin(qR) - qR * np.cos(qR) ) / (qR ** 3)  ) ** 2
```

```python
q = np.linspace(6E-3, 2, num=300)
R = 7
Int = Intensidade(q, R)
fig1, ax1 = plt.subplots(1, 1, figsize=(9,6))
fig2, ax2 = plt.subplots(1, 1, figsize=(9,6))
ax1.plot(q, Int/Int.max())
ax1.set(yscale='log', ylabel = 'Intensidade de espalhamento normalizada', xlabel='q/nm', ylim=(1E-5, 1.1), xlim=(0, 2))

ax2.plot(q * R, Int/Int.max())
ax2.set(yscale='log', ylabel = 'Intensidade de espalhamento normalizada', xlabel='qR', ylim=(1E-5, 1.1), xlim=(0, 14))
fig2.savefig(r'../saxs/espalhamento_esfera_monodispersa.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sp

def I_esfera (q, R):
    qR = q * R
    return ( 3 * ( np.sin(qR) - qR * np.cos(qR) ) / (qR ** 3)  ) ** 2

q = np.linspace(6E-3, 2, num=300)
R = 7
Int = I_esfera(q, R)
fig, ax = plt.subplots(1, 1, figsize=(9,6))

ax.plot(q * R, Int/Int.max())
ax.set(yscale='log', ylabel = 'Intensidade de espalhamento normalizada', xlabel='qR', ylim=(1E-5, 1.1), xlim=(0, 14))
```
