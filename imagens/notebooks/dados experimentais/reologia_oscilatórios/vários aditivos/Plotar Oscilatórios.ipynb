```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
import numpy as np
import matplotlib as mpl
import os
```

```python
files = glob.glob('*.csv')
files_pd = []
for file in files:
    temp_pd = pd.read_csv(file, encoding='utf8', delimiter=';', decimal=',')
    files_pd.append(temp_pd)
```

```python
for file in files_pd:
    wnum = 0
    G1num = 0
    G2num = 0
    
    wname = ''
    G1name = ''
    G2name = ''
    for i, col in enumerate(file.columns):
        if 'w' in col:
            wnum = i
            wname = col
        elif 'G1' in col:
            G1num = i
            G1name = col
        elif 'G2' in col:
            G2num = i
            G2name = col
    
    fig = plt.figure(figsize=(6,4))
    plt.plot(file[wname], file[G1name], marker='o', c='blue', label=G1name.replace('G1',"G'"), linewidth=0)
    plt.plot(file[wname], file[G2name], marker='o', markeredgecolor='red', label=G2name.replace('G2',"G''"), linewidth=0,
             markerfacecolor='w')
    plt.title(G1name.split(' ')[0])
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('$\omega$/rad.s$^{-1}$')
    plt.ylabel("G'/Pa, G''/Pa")
    plt.legend()
    fig.gca().tick_params(reset=True, axis='both', which='major', direction='out')
    
    counter = 1
    filename = G1name.split(' ')[0] + '.png'
    
    while os.path.isfile(filename):
        counter += 1
        filename = filename[:-4] + '--' + str(counter) + '.png'
    
    plt.savefig(filename, dpi=300)
    
    
```

## Água

```python
nomes = ['Agua', 'DMSO', 'Glic', 'Sac', 'Ur', '13BD']

files_a = glob.glob('*Agua*.csv')

files_pd = []

for file in files_a:
    temp_pd = pd.read_csv(file, encoding='utf8', delimiter=';', decimal=',')
    files_pd.append(temp_pd)

fig = plt.figure(figsize=(6,4))

markers = ['o', 's', '<', '>', '^', 'v']
colors = ['r', 'g', 'b', 'c', 'magenta']

markers.reverse()
colors.reverse()

for file in files_pd:
    
    wnum = 0
    G1num = 0
    G2num = 0
    
    wname = ''
    G1name = ''
    G2name = ''
    for i, col in enumerate(file.columns):
        if 'w' in col:
            wnum = i
            wname = col
        elif 'G1' in col:
            G1num = i
            G1name = col
        elif 'G2' in col:
            G2num = i
            G2name = col
    
    marker = markers.pop()
    color = colors.pop()
    
    plt.plot(file[wname], file[G1name], marker=marker, c=color, label=G1name.replace('G1',"G'"), linewidth=0)
    plt.plot(file[wname], file[G2name], marker=marker, markeredgecolor=color, label=G2name.replace('G2',"G''"), 
             linewidth=0, markerfacecolor='w')

plt.title('Água')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$\omega$/rad.s$^{-1}$')
plt.ylabel("G'/Pa, G''/Pa")
plt.legend()
fig.gca().tick_params(reset=True, axis='both', which='major', direction='out')

plt.savefig('Agua.png', dpi=300)
    
```
