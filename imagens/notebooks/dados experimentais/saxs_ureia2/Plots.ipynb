```python
mpl.rcParams.update({'font.size': 14, 'text.usetex':False, 'text.latex.unicode':True})

stylesC = ['-', '--', '-.', ':']
stylesC.reverse()
fig1, ax1 = plt.subplots(1, 1, figsize=(9,6))

def rename_surfactant(string):
    return string.replace('TTAB', r'C$_{14}$TAB').replace('CTAB', r'C$_{16}$TAB')

for i, file in enumerate(liquidos):
    df = pd.read_csv(file, sep=r'\s+', names=['q', 'I'], engine='python')
    df_m = df[df['q'] < 0.7]
    df_m2 = df[df['q'] > 0.9]
    ident = file.split('_')[0]
    if 'CTAB' in ren_dict[ident]:
        ax1.plot(df['q'], df['I'], label=rename_surfactant(ren_dict[ident]), c='C0', linestyle=stylesC.pop())
        
        Im = df_m['I'].max()
        m_x = df_m['I'].idxmax()
        qm = df_m['q'][m_x]
        
        Im2 = df_m2['I'].max()
        m_x2 = df_m2['I'].idxmax()
        qm2 = df_m2['q'][m_x2]
        
        ax1.annotate(s='$q_\mathrm{corr}$', xy=(qm, Im), xytext=(qm, Im*0.6), xycoords='data', 
                     arrowprops=dict(arrowstyle='-|>', color='k'))
        
        raio = 2 * np.pi/qm2
        t_raio = f'{raio:.2f}'
        ax1.annotate(s='$q_\mathrm{estr}$', xy=(qm2, Im2), xytext=(qm2, Im2*0.6), xycoords='data', 
                     arrowprops=dict(arrowstyle='-|>', color='r'), color='r')
        #ax1.text(qm2, Im2, f'{}')

ax1.set(xscale='log', yscale='log', xlabel='q/nm$^{-1}$', 
ylabel='Intensidade de espalhamento/u.a.', xlim=(0.1, 5.5))
ax1.xaxis.set_tick_params(which='major', width=1, length=10)
ax1.xaxis.set_tick_params(which='minor', width=1, length=4)
ax1.yaxis.set_tick_params(which='major', width=1, length=10)
ax1.yaxis.set_tick_params(which='minor', width=1, length=4)
ax1.legend()
# ax1.set_title('40% (m/m) ureia, CTAB', loc='right')
fig1.savefig(tese_path + r'\CTAB_q_corrs.pdf')
```

```python
mpl.rcParams.update({'font.size': 14, 'text.usetex':False, 'text.latex.unicode':True})

stylesC = ['-', '--', '-.', ':']
stylesT = ['-', '--', '-.', ':']
stylesD = ['-', '--', '-.', ':']
stylesC.reverse()
stylesD.reverse()
stylesT.reverse()
fig1, ax1 = plt.subplots(1, 1, figsize=(9,6))
lw = 2

# axins = inset_axes(ax1,
#                    width=1.5,  # width = 30% of parent_bbox
#                    height=1.,  # height : 1 inch
#                    loc=(0.1, 0.9))

axins = plt.axes([.17, .65, .2, .2])

lines_ctab = []
lines_ttab = []
lines_dtab = []

for i, file in enumerate(solidos):
    df = pd.read_csv(file, sep=r'\s+', names=['q', 'I'], engine='python')
    ident = file.split('_')[0]
    if 'b' in ren_dict[ident]:
        continue
    if 'CTAB' in ren_dict[ident]:
        c = ax1.plot(df['q'], df['I'], label=ren_dict[ident], 
                        c='C0', linestyle=stylesC.pop(), linewidth=lw)
        lines_ctab.extend(c)
        axins.plot(df['q'], df['I'], label=ren_dict[ident], 
                        c='C0', linestyle='-', linewidth=lw)
    if 'TTAB' in ren_dict[ident]:
        t = ax1.plot(df['q'], df['I'], label=ren_dict[ident], 
                        c='C1', linestyle=stylesT.pop(), linewidth=lw)
        lines_ttab.extend(t)
        axins.plot(df['q'], df['I'], label=ren_dict[ident], 
                        c='C1', linestyle='-', linewidth=lw)
    if 'DTAB' in ren_dict[ident]:
        d = ax1.plot(df['q'], df['I'], label=ren_dict[ident], 
                        c='C2', linestyle=stylesD.pop(), linewidth=lw)
        lines_dtab.extend(d)
        axins.plot(df['q'], df['I'], label=ren_dict[ident], 
                        c='C2', linestyle='-', linewidth=lw)

ax1.set(yscale='log', xlabel='q/nm$^{-1}$', 
        ylabel='Intensidade de espalhamento/u.a.', xlim=(1,5))

axins.set(yscale='log', xlim=(0, 2))
for tick in axins.xaxis.get_major_ticks():
    tick.label.set_fontsize(8)
for tick in axins.yaxis.get_major_ticks():
    tick.label.set_fontsize(8)
# axins.set_yticklabels([])
ax1.xaxis.set_tick_params(which='major', width=1, length=10)
ax1.xaxis.set_tick_params(which='minor', width=1, length=4)
ax1.yaxis.set_tick_params(which='major', width=1, length=10)
ax1.yaxis.set_tick_params(which='minor', width=1, length=4)
ax1.legend(title='Surfactantes sólidos', labels=['C$_{16}$TAB', 'C$_{14}$TAB', 'C$_{12}$TAB'])
# ax1.set_title('40% (m/m) ureia, CTAB', loc='right')
#ax1.set_title('Surfactantes sólidos', loc='right')
fig1.savefig(tese_path + r'\surfactante_solido.pdf')
```
